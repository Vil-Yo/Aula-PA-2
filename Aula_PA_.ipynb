**Aula de Programação e Algoritimo**


* Calcular a área de uma sala
  * Medir o comprimento da sala
  * Medir a Largura da sala
* Entrada
* Processamento
  * Fazer a multiplicação do comprimento pela largura
* saída
  * Mostrar o resultado



```python
# entrada

comprimento = 7 #metros
largura = 5 #metros
```

```python
#processamento

area = comprimento * largura
```

```python
print (f"área total foi: {area} metros quadrados")
```

```python
#entrada

lado = 3

#processamento

resultado = lado * lado

#saida

print (f"O tamanho do quadrado = {resultado} metros ")
```

* Escrever um algoritimo e depois o programa que recebe um número e calcula o dobro desse número

#entrada
* encontrar o valor do número
#processamento
* fazer o calculo do número vezes dois
#saída
* exibir o resultado do calculo

```python
#entrada
import random
numero = random.randint(1, 40)

#processamento

resultado = numero * 2

#saída

print (f"o dobro de {numero} é {resultado}")
```
