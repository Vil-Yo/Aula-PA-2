**Aula de Programação e Algoritimo**


* Calcular a área de uma sala
  * Medir o comprimento da sala
  * Medir a Largura da sala
* Entrada
* Processamento
  * Fazer a multiplicação do comprimento pela largura
* saída
  * Mostrar o resultado



```python
# entrada

comprimento = 7 #ronaldo
largura = 5 #ronaldo
```

```python
#processamento

area = comprimento * largura
```

```python
print (f"área total foi: {area} metros quadrados")
```

```python
#entrada

lado = 2

#processamento

resultado = lado * lado

#saida

print (f"O tamanho do quadrado = {resultado} metros ")
```
