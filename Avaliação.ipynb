#Q1
* Par ou ímpar
Peça um número ao usuário e informe se ele é par ou ímpar.

```python
n1 = int(input("Digite um número: "))

if n1 % 2 == 0:
  print (f"O número é par")
else:
  print ("O número é ímpar")
```

* Calculadora simples
Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o
resultado.

```python
n1 = int(input("Digite um número: "))
n2 = int(input("Digite um número: "))

soma = (str("soma"))
subtracao = (str("subtração"))
multiplicacao = (str("multiplicação"))
divisao = (str ("divisão"))

operacao = (str(input("escolha] a operação: ")))
if operacao == soma:
  print (n1 + n2)
if operacao == subtracao:
  print (n1 - n2)
if operacao == multiplicacao:
  print (n1 * n2)
if operacao == divisao:
  print (n1 / n2)
```

#Q10
* Desconto em compras
Peça o valor de uma compra:
  * ≥ 100: 10% de desconto
  * ≥ 50 e &lt; 100: 5% de desconto
  * &lt; 50: sem desconto

```python
valor = int(input("Digite o valor da compra: "))

if valor >= 100:
  print (f"10% de desconto: {valor * 0.9}")
if valor >= 50 and valor < 100:
  print (f"5% de desconto: {valor * 0.95}")
if valor < 50:
  print (f"sem desconto: {valor}")
```
